Analysis

```python
gdpValues[[ "Country", "y2023"]].sort_values(by=["y2023"], ascending=False).head(5)
```

```python
gdpValues[[ "Country", "y2033"]].sort_values(by=["y2033"], ascending=False).head(5)
```

```python
plt.plot(gdpValues.columns[1:], np.delete(gdpValues.loc[gdpValues["Country"] == "Germany"].values, 0))
plt.plot(gdpValues.columns[1:], np.delete(gdpValues.loc[gdpValues["Country"] == "India"].values, 0))
plt.show()
```

```python
plt.plot(gdpValues.columns[1:], np.delete(gdpValues.loc[gdpValues["Country"] == "United States"].values, 0))
plt.plot(gdpValues.columns[1:], np.delete(gdpValues.loc[gdpValues["Country"] == "China"].values, 0))
plt.show()
```

```python
yearList = gdpValues.columns[1:].to_numpy()
top5List = []
for i in range(len(yearList)):
    currentTop5 = gdpValues[["Country", yearList[i]]].sort_values(by=yearList[i], ascending=False).head(5)
    top5List.append([yearList[i].replace('y', ''), currentTop5["Country"].to_numpy()[0], currentTop5["Country"].to_numpy()[1], currentTop5["Country"].to_numpy()[2], currentTop5["Country"].to_numpy()[3], currentTop5["Country"].to_numpy()[4]])
top5DF = pd.DataFrame(top5List).transpose()
top5DF
```

```python
bottom5List = []
for i in range(len(yearList)):
    currentBottom5 = gdpValues[["Country", yearList[i]]].sort_values(by=yearList[i], ascending=False).tail(5)
    bottom5List.append([yearList[i].replace('y', ''), currentBottom5["Country"].to_numpy()[4], currentBottom5["Country"].to_numpy()[3], currentBottom5["Country"].to_numpy()[2], currentBottom5["Country"].to_numpy()[1], currentBottom5["Country"].to_numpy()[0]])
bottom5DF = pd.DataFrame(bottom5List).transpose()
bottom5DF
```

```python
countryList = gdpValues["Country"].to_numpy()
gdpDifferenceList = gdpValues["y2023"]- gdpValues["y2000"].to_numpy()

jointList = [countryList, gdpDifferenceList]
jointListDF = pd.DataFrame(jointList).transpose()
jointListDF = jointListDF.sort_values(by=1, ascending=False)
print(jointListDF.head())
print(jointListDF.tail())
```
